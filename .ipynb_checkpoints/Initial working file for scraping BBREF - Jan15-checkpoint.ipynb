```python
import requests
```

```python
url = 'https://www.baseball-reference.com/teams/TBR/2018-schedule-scores.shtml'

response = requests.get(url)
```

```python
response.status_code
```

```python
print(response.text)
```

```python
page = response.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
"""
BELOW IS THE CODE FOR ADJACENT SEASONS - COME BACK TO THIS TO GET MULTIPLE YEARS
       <a class="button2 prev" href="/teams/TBR/2013-schedule-scores.shtml">
        2013 Season
       </a>
       <a class="button2 next" href="/teams/TBR/2015-schedule-scores.shtml">
        2015 Season
"""
```
